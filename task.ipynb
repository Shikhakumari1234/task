```python
# import all imp libraries for implementation
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.manifold import TSNE
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
#load the dataset for the implementation
iris = load_iris() #i have use in-built dataset of the sklearn libarary , we can also use any dataset presnct in excel or csv format
X = iris.data  # Features
y = iris.target  # Target
```

```python
#prepare the data 
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
k = 10  # defining the Number of neighbors for k-NN
neigh = NearestNeighbors(n_neighbors=k)
neigh.fit(X_scaled)
knn_indices = neigh.kneighbors(return_distance=False)

```

```python
#apply t-SNE functions
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
X_tsne = tsne.fit_transform(X_scaled)
```

```python
#plotting the graph
plt.figure(figsize=(8, 6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, cmap=plt.cm.Spectral)
plt.title('t-SNE Visualization')
plt.savefig('tSNE_visualization.png')
```

```python
#check the accuracy
knn_accuracy = np.mean(knn_indices[:, 0] == knn_indices[:, 1])

print(f'k-NN Accuracy: {knn_accuracy:.2f}')
```

```python
#A k-NN Accuracy of 0.00 indicates that none of the nearest neighbors in the k-NN sampling are the same as the points they were sampled from.
```

```python
# graphh for the knn sampling
plt.figure(figsize=(8, 6))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c='blue', s=20)  # All points in blue
plt.scatter(X_tsne[knn_indices, 0], X_tsne[knn_indices, 1], c='red', s=40, marker='x')  # k-NN samples in red
plt.title('k-NN Sampling')
plt.savefig('kNN_sampling.png')
```

```python
# plotting for the dyanamic knn 
tsne_dynamic = TSNE(n_components=2, perplexity=30, random_state=42)
X_tsne_dynamic = tsne_dynamic.fit_transform(X_scaled)

# Step 4: Visualize t-SNE results for dynamic k-NN
plt.figure(figsize=(8, 6))
plt.scatter(X_tsne_dynamic[:, 0], X_tsne_dynamic[:, 1], c=y, cmap=plt.cm.Spectral)
plt.title('Dynamic k-NN t-SNE Visualization')
plt.savefig('dynamic_kNN_tSNE_visualization.png')
```

```python
knn_accuracy_dynamic = np.mean(knn_indices[:, 0] == knn_indices[:, 1])  # Modify this based on dynamic k-NN

print(f'Dynamic k-NN Accuracy: {knn_accuracy_dynamic:.2f}')
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_tsne_dynamic[:, 0], X_tsne_dynamic[:, 1], c='blue', s=20)  # All points in blue
plt.scatter(X_tsne_dynamic[knn_indices, 0], X_tsne_dynamic[knn_indices, 1], c='red', s=40, marker='x')  # k-NN samples in red
plt.title('Dynamic k-NN Sampling')
plt.savefig('dynamic_kNN_sampling.png')
```

```python
perplexity = 5

k=10
# Step 3: Apply t-SNE for random Sampling
random_indices = np.random.choice(len(X_scaled), k, replace=False)
X_random = X_scaled[random_indices]
y_random = y[random_indices]  # Select corresponding target values

# Step 3: Apply t-SNE for random Sampling
tsne_random = TSNE(n_components=2, perplexity=perplexity, random_state=42)
X_tsne_random = tsne_random.fit_transform(X_random)

# Step 4: Visualize t-SNE results for random Sampling
plt.figure(figsize=(8, 6))
plt.scatter(X_tsne_random[:, 0], X_tsne_random[:, 1], c=y_random, cmap=plt.cm.Spectral)
plt.title('t-SNE Visualization for Random Sampling')
plt.savefig('random_sampling_tSNE_visualization.png')
```
